```python
# Pegando apenas os cod_mun, nome_mun,cod_uf e cod_regiao
municipios_separacao = municipios[['cod_mun','nom_mun','cod_uf','cod_regiao']]
# Pegando apenas os cod_mun, nome_mun,cod_uf e cod_regiao
municipios_separacao = municipios[['cod_mun','nom_mun','cod_uf','cod_regiao']]
# tabela com nomes dos estados por codigo
estados = pd.read_csv('Dados/estados_com_regiao.csv')
# mudando o nome da coluna codigo_uf para cod_uf
estados = estados.rename(columns={'codigo_uf':'cod_uf'})
# junção das tabelas municipios_separacao e estados
municipios_separacao = municipios_separacao.merge(estados, on='cod_uf')
# separando cod_regiao e nom_regiao
regiao = municipios_separacao[['cod_regiao','regiao']]
regiao = regiao.drop_duplicates()
display(regiao)
```

3. Aqui eu estou tratando a base Que agrega por municipio para colocar os dados das regioes e o nome dos estados

```python
# colocando o nome da regiao na tabela municipios pelo cod_regiao
municipios_dois = municipios.merge(regiao, on='cod_regiao')
# trocando cod_regiao por reagiao
municipios_dois = municipios_dois.rename(columns={'regiao':'nome_regiao'})
# colocando uf na tabela municipios pelo cod_uf
municipios_dois = municipios_dois.merge(estados, on='cod_uf')
# tirando colunas desnecessarias da tabela municipios_dois latitute e longitude regiao
municipios_dois = municipios_dois.drop(columns=['latitude','longitude','regiao'])
# trocando coluna nome por nome_uf
municipios_dois = municipios_dois.rename(columns={'nome':'nome_uf'})
display(municipios_dois)
```

4. Deletando coulnas inuteis

Vou utilizar os dados agrupados pobres e vulneraveis juntos as que tem a tah _pob_vul

```python
municipios_dois.columns
```

```python
# deletando campos desnecessarios
municipios_tratado = municipios_dois.drop(columns=['qtd_dom_pobres', 'qtd_dom_vulneraveis','IPM_quintis', 'IPM_quintis_regiao',
       'IPM_quintis_uf', 'qtd_pes_pobres', 'qtd_pes_vulneraveis','qtd_0a11_pobres', 'qtd_0a11_vulneraveis','cod_mun', 'cod_regiao', 'cod_uf','nome_uf','num_ibeu','incidencia','contr_E1', 'contr_E2', 'contr_E3', 'contr_S1',
       'contr_S2', 'contr_S3', 'contr_S4', 'contr_T1', 'contr_T2', 'contr_T3',
       'contr_P1', 'contr_P2', 'contr_P3', 'contr_P4'],axis=1)
# dados numericos com no maximo duas casas decimais
pd.options.display.float_format = "{:.2f}".format
display(municipios_tratado)
```
